```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle


import pandas as pd
import numpy as np
```

```python

# Load model
model = load_model('model.h5')

# Load one-hot encoder
with open('onehotEncoder_geography.pkl', 'rb') as file:
    label_onehotEncoder = pickle.load(file)

# Load scaler
with open('scale_data.pickle', 'rb') as file:
    scaler = pickle.load(file)
    
# label encoder 
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder=pickle.load(file)
```

```python
raw_input = {
    "CreditScore": 650,
    "Geography": "France",
    "Gender": "Male",
    "Age": 40,
    "Tenure": 5,
    "Balance": 60000,
    "NumOfProducts": 2,
    "HasCrCard": 1,
    "IsActiveMember": 1,
    "EstimatedSalary": 50000
}
input_df = pd.DataFrame([raw_input])
```

```python
geo_encoded = label_onehotEncoder.transform(input_df[['Geography']])
geo_df = pd.DataFrame(
    geo_encoded,
    columns=label_onehotEncoder.get_feature_names_out(['Geography'])
)

input_df = pd.concat([input_df.drop('Geography', axis=1), geo_df], axis=1)

# 3. Encode Gender (same encoder used in training)
input_df['Gender'] = label_encoder.transform(input_df['Gender'])
```

```python
input_df
```

```python
X_scaled = scaler.transform(input_df)
```

```python
prediction = model.predict(X_scaled)

print("Churn probability:", prediction[0][0])
print("Churn?" , "Yes" if prediction[0][0] > 0.5 else "No")
```
